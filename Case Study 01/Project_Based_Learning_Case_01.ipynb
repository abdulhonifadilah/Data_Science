```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
data = [100,150,50,100,130,120,100,85,70,150,150,120,50,100,100,140,90,150,50,90,120,100,110,75,65]
df = pd.DataFrame(data, columns=["TransactionAmount"])
print(df)
```

```python
#Central Tendency
import statistics
print("mean : ", statistics.mean(data))
#print median
print("median : ", statistics.median(data))
```

#analisis
1. rata-rata penjualan setelah training adalah 102.6 mengalami peningkatan sebesar 2.6 yang awalnya adalah 100
2. nilai mediannya adalah 100

```python
#Variability
#import numpy as np
print(np.var(data))
print(np.std(data))
print(np.max(data) - np.min(data))
print(np.quantile(data, 0.25))
print(np.quantile(data,0.5))
print(np.quantile(data, 0.75))
```

analisis
1. Variance yang didapatkan adalah 972.2399999999999
2. Standard deviansi adalah 31.18076330047101
3. Range adalah 100
4. Q1 = 85 Q2 = 100 dan Q3 = 120

```python
#Define H0 and H1
#H0 = rata-rata sebelum training 100
#H1 = rata-rata setelah training > 100
```

```python
#Define alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test
stats.ttest_1samp(a=data, popmean=100)
```

```python
#statistic= 0.40 , pvalue= 0.68
```

```python
#Conclude with alpha and p-value
#Conclusion: p-value jauh dari alpha maka gagal menolak H0
#Training tidak efektif karena rata-rata setelah training masih di nilai 100
```

```python
#Conclude with t statistics and critical region
#Find the critical region
```

```python
#H1: larger(right side)
stats.t.ppf(1-0.05, len(data)-1)
```

```python
#Critical region result = 1.7
# nilai statistic di luar critical region
#Conclusion: Training tidak efektive,rata-rata penjualan masih dinilai 100.
```
